```python
import pandas as pd
```

```python
data = pd.read_csv('Online_Retail.csv', encoding='windows-1252')
```

```python
data
```

```python
data_uk = data[data.Country=='United Kingdom']
```

```python
data_uk = data_uk.dropna()
```

```python
data_uk['Quantity'] = data_uk['Quantity'][data_uk['Quantity'] >= 0]
```

```python
data_uk = data_uk.dropna()
```

```python
data_uk.isnull().sum()
```

```python
data_uk = data_uk[~data_uk.InvoiceNo.str.contains('C',na=False)]
```

```python
data_uk.duplicated().sum()
```

```python
data_uk = data_uk.drop_duplicates(keep="first")
```

```python
products_quantity = data_uk.groupby('Description')['Quantity'].sum().reset_index()
```
